```python
# definindo classes, poses, iluminações e fundos possíveis
classes = [
           'celular',
           'livro',
           'sapato',
           'chinelo',
           'caneca',
           'portacopo',
           'garrafa',
           'prato',
           'tesoura',
           'chave'
]

# as poses, fundos e objetos precisam ser constantes dentro de um objeto
# embora possam ser únicas entre objetos
poses = ['p1', 'p2', 'p3']
fundos = ['f1', 'f2', 'f3']
objetos = ['obj1', 'obj2', 'obj3', 'obj4']

# Dia Interno, Dia Externo, Noite Interno, Noite Externo
ilums = ['di', 'de', 'ni', 'ne']

# vamos adotar a seguinte nomenclatura para os arquivos de imagem:
# [classe]_[obj]_[pose]_[iluminacao]_[fundo]
# exemplo: para a classe 'celular', o arquivo que define a foto tirada
# com o primeiro objeto na pose normal, dentro de casa, de dia, com fundo azul
# seria: celular_obj1_p1_di_f1
```

```python
# funcoes auxiliares
import pandas as pd
from itertools import product
def gera_combinacoes(classe):
  out = []
  for obj, pose, ilum, fundo in product(objetos, poses, ilums, fundos):
    out.append(f'{classe}_{obj}_{pose}_{ilum}_{fundo}.png')
  return out
```

```python
# número de objetos por classe
n_objetos = 4
# número de poses para cada objeto
n_poses = 3
# número de variações para a iluminação
n_iluminacao = 4
# número de variações para o fundo
n_fundos = 3

# vamos checar que nosso dataset possui os registros necessários e registrar
# os nomes dos regitros a serem feitos no arquivo imgs_todo.csv
import os

overall_missing = []
for classe in classes:
  # cria a pasta caso ainda não exista
  os.makedirs(f'{workdir}/dataset/{classe}', exist_ok=True)
  # lista arquivos dentro da pasta
  objs = os.listdir(f'{workdir}/dataset/{classe}')
  # cria lista com todas as combinações para uma classe
  expected = gera_combinacoes(classe)
  # descobre quais estão faltando
  missing = [x for x in expected if x not in objs]
  overall_missing.extend(missing)

# criando txt com lista de arquivos faltantes
with open(f"{workdir}/dataset/imagens_faltando.txt", "w") as outfile:
  outfile.write("\n".join(overall_missing))

print(f'{len(overall_missing)} imagens a serem fotogradas.')
```
