
Calculando métricas de erro/desempenho das predições vs ground truth
---



```python
from sklearn.metrics import jaccard_similarity_score

def jaccard_index(img_a: np.array, img_b: np.array) -> float:
  if img_a.size != img_b.size:
    raise ValueError('Image sizes differ.')
  img_a_flat, img_b_flat = img_a.copy().flatten(), img_b.copy().flatten()
  return jaccard_similarity_score(img_a_flat, img_b_flat)

```
