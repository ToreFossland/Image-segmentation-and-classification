###1)256x256
###2)Gray

```python
import os
```

```python
### Obs: 'ep_dataset' was the alias I used for the folder name.
path_prefix = "/my-drive/MyDrive/ep_dataset/"
gray_dataset_path = path_prefix + "dataset_gray/"
```

```python
from skimage.color import rgb2gray
import imageio
import numpy as np
```

```python
okay_flag = True
with open(path_prefix + "sanity_check_log.txt", "w+") as log:
  log.write("Sanity check output:\n")
  for dirpath, dirname, filename in os.walk(gray_dataset_path):
    if dirpath != gray_dataset_path:
      for file in filename:
        if file.split(".")[1] in ["jpg", "png", "jpeg"]:
          gray_img = imageio.imread(dirpath + f"/{file}")
          if gray_img.shape != (256, 256):
            okay_flag = False
            message = f"File {dirpath}/{file} has shape {gray_img.shape} instead of (256, 256)!\n"
            log.write(message)
            print(message)
  if okay_flag:
    log.write("All files are in the correct format!\n")
```
